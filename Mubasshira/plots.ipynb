```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = './All Approx Results/minpath+local_track_C.csv'
output_path = './All Approx Results/minpath+local_trackC_plots/'
data = pd.read_csv(file_path)

data['Runtime (s)'] = data['Runtime (ms)'] / 1000

# Terminals vs Runtime (s)
plt.figure(figsize=(8, 6))
plt.scatter(data['Terminals'], data['Runtime (s)'], color='blue', label='Runtime(s)')
plt.xlabel('Terminals')
plt.ylabel('Runtime(s)')
plt.title('Terminals vs Runtime(s)')
plt.grid(True)
plt.savefig(f'{output_path}terminals_vs_runtime.png')
plt.close()

# Nodes vs Runtime (s)
plt.figure(figsize=(8, 6))
plt.scatter(data['Nodes'], data['Runtime (s)'], color='green', label='Runtime(s)')
plt.xlabel('Nodes')
plt.ylabel('Runtime(s)')
plt.title('Nodes vs Runtime(s)')
plt.grid(True)
plt.savefig(f'{output_path}nodes_vs_runtime.png')
plt.close()

# Edges vs Runtime (s)
plt.figure(figsize=(8, 6))
plt.scatter(data['Edges'], data['Runtime (s)'], color='red', label='Runtime(s)')
plt.xlabel('Edges')
plt.ylabel('Runtime(s)')
plt.title('Edges vs Runtime(s)')
plt.grid(True)
plt.savefig(f'{output_path}edges_vs_runtime.png')
plt.close()

# Terminals vs Ratio
plt.figure(figsize=(8, 6))
plt.scatter(data['Terminals'], data['Ratio'], color='purple', label='Approximation Ratio')
plt.xlabel('Terminals')
plt.ylabel('Approximation Ratio')
plt.title('Terminals vs Approximation Ratio')
plt.grid(True)
plt.savefig(f'{output_path}terminals_vs_ratio.png')
plt.close()

# Nodes vs Ratio
plt.figure(figsize=(8, 6))
plt.scatter(data['Nodes'], data['Ratio'], color='orange', label='Approximation Ratio')
plt.xlabel('Nodes')
plt.ylabel('Approximation Ratio')
plt.title('Nodes vs Approximation Ratio')
plt.grid(True)
plt.savefig(f'{output_path}nodes_vs_ratio.png')
plt.close()

# Edges vs Ratio
plt.figure(figsize=(8, 6))
plt.scatter(data['Edges'], data['Ratio'], color='cyan', label='Approximation Ratio')
plt.xlabel('Edges')
plt.ylabel('Approximation Ratio')
plt.title('Edges vs Approximation Ratio')
plt.grid(True)
plt.savefig(f'{output_path}edges_vs_ratio.png')
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

minpath_file = './All Approx Results/minpath_track_A.csv'  
minpath_local_file = './All Approx Results/minpath+local_track_A.csv' 
output_path =  './All Approx Results/minpath_vs_min+local/'

minpath_data = pd.read_csv(minpath_file)
minpath_local_data = pd.read_csv(minpath_local_file)

minpath_data['Runtime (s)'] = minpath_data['Runtime (ms)'] / 1000
minpath_local_data['Runtime (s)'] = minpath_local_data['Runtime (ms)'] / 1000

# Merge the dataframes on Test Case, keeping only instances present in both files
merged_data = pd.merge(minpath_data[['Test Case', 'Runtime (s)', 'Ratio']],
                       minpath_local_data[['Test Case', 'Runtime (s)', 'Ratio']],
                       on='Test Case',
                       how='inner',
                       suffixes=('_minpath', '_minpath_local'))


# Minpath vs Minpath+Local Runtimes
merged_data.sort_values(by='Runtime (s)_minpath_local', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Test Case'], merged_data['Runtime (s)_minpath'], marker='o', label='Minpath Runtime(s)', color='blue')
plt.plot(merged_data['Test Case'], merged_data['Runtime (s)_minpath_local'], marker='o', label='Minpath+Local Runtime(s)', color='purple')
plt.xlabel('Test Cases')
plt.ylabel('Runtime(s)')
plt.title('Minpath vs Minpath+Local Runtimes')
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.savefig(f'{output_path}minpath_vs_minpath_local_runtimes.png')
plt.close()

# Minpath vs Minpath+Local Ratios
merged_data.sort_values(by='Ratio_minpath_local', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Test Case'], merged_data['Ratio_minpath'], marker='o', label='Minpath Ratio', color='red')
plt.plot(merged_data['Test Case'], merged_data['Ratio_minpath_local'], marker='o', label='Minpath+Local Ratio', color='green')
plt.xlabel('Test Case')
plt.ylabel('Ratio')
plt.title('Minpath vs Minpath+Local Approximation Ratios')
plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.savefig(f'{output_path}minpath_vs_minpath_local_ratios.png')
plt.close()
```
